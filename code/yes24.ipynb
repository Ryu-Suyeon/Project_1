```python
# 파일에서 중복된 문장을 제거하여 새로운 파일에 저장하는 함수
def remove_duplicate_sentences(input_file, output_file):
    unique_sentences = set()

    # 파일 열기 및 중복 제거
    with open(input_file, 'r', encoding='utf-8') as file:
        for line in file:
            # 줄 바꿈 문자 제거 후 공백을 기준으로 문장 분리
            sentences = line.strip().split('.')
            for sentence in sentences:
                sentence = sentence.strip()
                if sentence:
                    unique_sentences.add(sentence + '.')

    # 중복 제거된 문장을 새 파일에 저장
    with open(output_file, 'w', encoding='utf-8') as file:
        for sentence in unique_sentences:
            file.write(sentence + '\n')

# 입력 파일과 출력 파일명 지정
input_filename = 'yes24reviews책이름.txt'  # 입력 파일명
output_filename = 'yes24reviews책이름remove.txt'  # 출력 파일명

# 함수 호출하여 중복 제거 후 새 파일에 저장
remove_duplicate_sentences(input_filename, output_filename)
```
