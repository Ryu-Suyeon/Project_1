```python
import pandas as pd
df_dic = pd.read_csv('./Korean_Emotion_POSWord_Dictionary.csv') # 감정 사전 csv 파일
print(df_dic[:100])
```

```python
#단어-감정 사전 만들기
wordEmotionDic = {}
for i in range(0,len(df_dic)):
  wordEmotionDic[df_dic.iloc[i,2]] = df_dic.iloc[i,-1]
print(wordEmotionDic)
```

```python
emotionCount = df_dic.groupby("대표감정")['단어'].count()
print(emotionCount)
emotions = list(emotionCount.index)
print(emotions)
```

```python
# 데이터 불러오기
import re

fileDir = "###/크롤링/yes24/yes24reviews아가씨.txt" # 파일 경로 수정
fr = open(fileDir, 'r', encoding='utf-8')
contents = fr.readlines()
fr.close()

reviewList = []
for content in contents:
  content = re.sub('[^가-힣]', ' ', content)
  content = re.sub('[\s]+', ' ', content).strip()
  reviewList.append(content)

for eachReview in reviewList:
  print(eachReview)
  
```

```python
from kiwipiepy import Kiwi

kiwi = Kiwi()
#품사태깅된 문장
reviewPOSList = []
#단어의 타입 확인
wordType = set()

for eachReview in reviewList:
  #띄어쓰기
  eachReviewSpace = kiwi.space(eachReview)
  #품사태깅
  posTagout = kiwi.tokenize(eachReviewSpace)
  sentencePOS = ''
  for eachOut in posTagout:
      result = eachOut[0]+"/"+eachOut[1]
      sentencePOS = sentencePOS + " " + result
      if result != "":
        wordType.add(result)
  sentencePOS = sentencePOS.strip()
  reviewPOSList.append(sentencePOS)
  print(sentencePOS)

print(wordType)
print(len(wordType))
print(len(reviewPOSList))
```

```python
import pandas as pd
reviewIndex = []
for i in range(0,len(reviewPOSList)):
  reviewIndex.append(str(i))
# 0으로 이루어진 단어 행렬 생성 (초기값)
matrixEmotion = pd.DataFrame(0.0, index=reviewIndex, columns=emotions)
print(matrixEmotion)
```

```python
import pandas as pd

# 주어진 문서 리스트와 단어-감정 사전
reviewIndex = []
for i in range(0, len(reviewPOSList)):
    reviewIndex.append(str(i))

# 초기값이 0.0인 문서 단어 행렬 생성
matrixEmotion = pd.DataFrame(0.0, index=reviewIndex, columns=emotions)

# 각 문서에서 단어가 등장할 때마다 행렬 값 업데이트
for i in range(len(reviewPOSList)):
    sentence = reviewPOSList[i]
    words = sentence.split(" ")
    for word in words:
        if word != '' and word in wordEmotionDic:
            matrixEmotion.loc[str(i), wordEmotionDic[word]] += 1.0

print(matrixEmotion)

df = pd.DataFrame(matrixEmotion)

# DataFrame을 엑셀 파일로 저장
excel_file_path = '아가씨_문서단어_yes24.xlsx'  # 파일명 수정
df.to_excel(excel_file_path, index=True, header=True)
```

```python
import pandas as pd

# 엑셀 파일 불러오기
file_path = '아가씨_문서단어_yes24.xlsx'  # 엑셀 파일 경로
df = pd.read_excel(file_path)

# 각 열의 값을 모두 더하여 새로운 행으로 추가
sum_values = df.sum()  # 각 열의 값을 더함
df_summed_row = pd.DataFrame(sum_values).T  # 새로운 행으로 추가하기 위해 DataFrame 생성 및 전치

# 기존 DataFrame을 새로운 행으로 대체
df = df_summed_row

# 결과 출력
print(df)

# DataFrame을 엑셀 파일로 저장
excel_file_path = '아가씨_감정합계_yes24.xlsx'  ##수정##
df.to_excel(excel_file_path, index=True, header=True)
```

```python
import pandas as pd

# 파일명 수정, 모든 리뷰를 각 사이트로 행을 만든 후, 하나의 파일로 저장
excel_files = ['아가씨_감정합계_yes24.xlsx', '아가씨_감정합계_교보문고.xlsx', '아가씨_감정합계_알라딘.xlsx', '아가씨_감정합계_네이버영화.xlsx']
new_index = ['yes24', '교보문고', '알라딘', '네이버영화']

# 각 엑셀 파일을 DataFrame으로 읽고 리스트에 저장
dfs = []
for excel_file in excel_files:
    df = pd.read_excel(excel_file)
    dfs.append(df)

# 데이터를 수직으로 병합하기
total_df = pd.concat(dfs, ignore_index=True)

# 2번째, 3번째 열 삭제하기
total_df = total_df.drop(columns=[total_df.columns[0], total_df.columns[1]])

# 인덱스 변경하기
total_df.index = new_index

print(total_df)
excel_file_path = '아가씨_감정_통합.xlsx'  # 저장할 엑셀 파일 경로와 파일명
total_df.to_excel(excel_file_path, index=True, header=True)
```
